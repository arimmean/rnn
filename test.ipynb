```python
import os
import torch
from typing import Union, List, Tuple
from sentencepiece import SentencePieceTrainer, SentencePieceProcessor
from torch.utils.data import Dataset
from torch import default_generator, Generator, randperm, Tensor
```

```python
data_file = 'jokes.txt'

with open(data_file) as file:
            texts = file.readlines()

gen = torch.Generator().manual_seed(42)
train_subset, val_subset = torch.utils.data.random_split(texts, [1 - 0.05, 0.05], generator=gen)
```

```python
SentencePieceTrainer.train(
                input=data_file, vocab_size=2000,
                model_type='bpe', model_prefix="sp_model_prefix",
                normalization_rule_name="nmt_nfkc_cf"
            )
sp_model = SentencePieceProcessor(model_file="sp_model_prefix" + '.model')
```

```python
indices = sp_model.encode(list(map(texts.__getitem__, train_subset.indices)))
```
